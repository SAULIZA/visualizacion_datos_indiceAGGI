```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
filename = "/content/drive/MyDrive/Global Radiative Forcing - 1979-2021.xlsx"
df = pd.read_excel(filename)
df.head(22)
```

```python
df.info()
```

```python
plt.style.use("seaborn")
```

```python
#transformamos las columnas del data frame a arrays de numpy
Year = df["Year"].values
CO_2 = df["CO2"].values
CH_4 = df["CH4"].values
N_2O = df["N2O"].values
CFC  = df["CFCs"].values
HCFC = df["HCFCs"].values
HFC = df["HFCs"].values
SUM = df["Total"].values
```

```python
len(Year)
```

```python
len(SUM)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols= 3,figsize=(15,5))
axes[0].plot(Year, SUM, c="y", label ="AGGI total index")
axes[0].plot(Year, CO_2, c="k", label ="AGGI index for CO2")
axes[0].legend(fontsize=10)
axes[0].set_xlabel("Year", fontsize=12)
axes[0].set_ylabel("AGGI index", fontsize=12)

axes[1].plot(Year, CH_4, c="b", label ="AGGI index for CH4")
axes[1].plot(Year, N_2O, c="g", label ="AGGI index for N2O")
axes[1].plot(Year, CFC, c="r", label ="AGGI index for CFCs")
axes[1].legend(loc=(0.4,0.5), fontsize=10)
axes[1].set_xlabel("Year", fontsize=12)
axes[1].set_ylabel("AGGI index", fontsize=12)

axes[2].plot(Year, HCFC, c="c", label ="AGGI index for HCFCs")
axes[2].plot(Year, HFC, c="m", label ="AGGI index for HFCs")
axes[2].legend(loc="best", fontsize=10)
axes[2].set_xlabel("Year", fontsize=12)
axes[2].set_ylabel("AGGI index", fontsize=12)
#axes.set_xlabel("Year", fontsize=16)
#axes.set_ylabel("AGGI index", fontsize=16)
#axes.legend(loc=(1.02,0.6));
```

Los índices de gases de efecto invernadero anuales (AGGI) de los gases HFCSs Y HCFCs han crecido durante todo el tiempo, sin embargo sus niveles son muy bajos. Es muy interesante lo que sucede con la evolución del índice AGGI de las sustancias CFCs, estas fueron prohibidas ya que su uso perjudicaba notablemente la capa de ozono, y en efecto se puede ver que el índice de esta sustacia tiende a disminuir por encima de 1992, lo cual es lógico ya que el índice de AGGI esta relacionado con la concentración de la sustancia. Por otro lado, los índices de los gases CH4 y N2O se encuentran en constante aumento, y sus valores son relativamente bajos. El índice del gas CO2 tiene una contribución importante al índice total, ya que sus valores son los más altos comparado con el resto de sustancias, y su incremento es notablemente mayor que cualquier otro gas de efecto invernadero, seguramente este aumento se debe a la gran contaminación que provienen del uso de combustibles fósiles y la tala de arboles que aumenta con el transcurso del tiempo.
